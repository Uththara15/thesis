```python
import spacy

for model_name in ["fi_core_news_sm", "en_core_web_sm"]:
    try:
        nlp = spacy.load(model_name)
        meta = nlp.meta
        print(f"{model_name}")
        print(f"  version : {meta.get('version')}")
        print(f"  spaCy   : {meta.get('spacy_version')}")
        print(f"  lang    : {meta.get('lang')}")
        print()
    except Exception as e:
        print(f"{model_name}: NOT LOADABLE ({e})")
        print()
```

```python
import pandas as pd
from pathlib import Path

# Adjust path if your notebook is not in the project root
PROJECT_ROOT = Path.cwd()
ict_path = PROJECT_ROOT / "data" / "raw" / "final_dataset.csv"
print(f"Loading: {ict_path}")
print(f"Exists : {ict_path.exists()}")

df = pd.read_csv(ict_path)
print(f"\nShape  : {df.shape}")
print(f"\nColumns:\n{df.columns.tolist()}")

print(f"\n--- similarity_label value counts ---")
print(df['similarity_label'].value_counts())

print(f"\n--- pair_type value counts ---")
print(df['pair_type'].value_counts())

print(f"\n--- pair_difficulty value counts ---")
print(df['pair_difficulty'].value_counts())

print(f"\n--- pair_type x similarity_label cross-tab ---")
print(df.groupby(['pair_type', 'similarity_label']).size())

print(f"\n--- pair_difficulty x similarity_label cross-tab ---")
print(df.groupby(['pair_difficulty', 'similarity_label']).size())

print(f"\n--- pair_type x pair_difficulty cross-tab ---")
print(df.groupby(['pair_type', 'pair_difficulty']).size())
```

```python
import pandas as pd
from pathlib import Path

PROJECT_ROOT = Path.cwd()
df = pd.read_csv(PROJECT_ROOT / "data" / "raw" / "final_dataset.csv")

text_cols = [c for c in ['title_fi', 'title_en',
                         'outcomes_fi', 'outcomes_en',
                         'contents_fi', 'contents_en',
                         'assessment_fi', 'assessment_en'] if c in df.columns]

print("Field non-null counts and character-length statistics:\n")
for col in text_cols:
    series = df[col].dropna().astype(str)
    char_lens = series.str.len()
    word_lens = series.str.split().str.len()
    print(f"{col:<18}  n_filled={len(series):>4}  "
          f"chars: mean={char_lens.mean():>7.0f} median={char_lens.median():>6.0f} max={char_lens.max():>6}  "
          f"words: mean={word_lens.mean():>5.0f} median={word_lens.median():>5.0f} max={word_lens.max():>5}")
```
